# How to find Data set on Internet download,import in csv with Pandas and visualize it

https://datahub.io/core/co2-ppm : c02-mm-mlo

```python
import pandas as pd
```

```python
pd.read_csv("https://datahub.io/core/co2-ppm/r/co2-mm-mlo.csv")  #right click copy link from site ""
```

```python
df = pd.read_csv("https://datahub.io/core/co2-ppm/r/co2-mm-mlo.csv")
```

```python
df["Date"] # object data type, when numpy doesnt recognise it, just storing as array of string
# https://pandas.pydata.org/docs/user_guide/timeseries.html
```

```python
pd.to_datetime(df["Date"])    # passing in above, dtype changes
```

```python
df["Datetime"] = pd.to_datetime(df["Date"])  # adding new col
```

```python
df   # date is string, datetime is datetime
```

# Seaborn

```python
import seaborn as sns   
# https://seaborn.pydata.org/tutorial/relational.html
# Homehttps://seaborn.pydata.org/generated/seaborn.lineplot.html
```

```python
sns.lineplot(data=df, x="Datetime", y="Interpolated")
sns.lineplot(data=df, x="Datetime", y="Trend")
```
