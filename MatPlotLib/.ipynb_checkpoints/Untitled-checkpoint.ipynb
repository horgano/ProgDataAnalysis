# 2 Plots/ 1 Axis

```python
x = np.arange(0.0, 10.0, 0.01)
y = 3 * x
noise = np.random.normal(0.0, 2.0, len(x))
```

```python
plt.plot(x, y, 'b.')
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot(x, y, 'b.')
plt.plot(x, noise, 'g-')
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot(x, y, 'b.')
plt.plot(x, y + noise, 'g-')
plt.ylabel('some numbers')
plt.show()
```

# Graph Annotation

```python
plt.plot(x, y + noise, 'g-', label= "Actual Data")
plt.plot(x, y, 'b.', label="Fitted Model")

plt.title("Average speed vs distance covered over time")
plt.xlabel("Average Speed (m/s)")
plt.ylabel('Distance covered (km)')
plt.legend()

plt.show()
```

# Histograms

# Side by side plots

```python
# rows, columns instances, 
a = np.random.normal(0.0, 1.0, 100000)
plt.subplot(1, 2, 1)
plt.hist(a, bins = 80)
plt.show()

a = np.random.uniform(-4.0, 3.0, 1000)
plt.subplot(1, 2, 1)
plt.hist(a)
plt.show()
```

```python
# rows, columns instances, 
a = np.random.normal(0.0, 1.0, 100000)
plt.subplot(2, 1, 1)
plt.hist(a, bins = 80)
plt.show()

a = np.random.uniform(-4.0, 3.0, 1000)
plt.subplot(2, 1, 1)
plt.hist(a)
plt.show()
```

# Other Plots

```python
c = np.arange(-2.0 * np.pi, 2.0 * np.pi, 0.1)

plt.plot(c, np.sin(c), 'g.')
plt.plot(c, np.cos(c), 'b-')

plt.show()
```

```python
d = np.arange(1.0, 25.0, 0.1)

plt.plot(d, d**2, 'g-', label='d^2')
plt.plot(d, d**3, 'b-', label='d^3')
plt.plot(d, d**4, 'r-', label='d^4')
plt.plot(d, d**2, 'y-', label='d^2')

plt.legend()
plt.show()
```
