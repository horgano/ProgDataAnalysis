# Classification

# Userguide 1.6: Nearest Neighbours
Instance based learning or non-generalizing, n neighbours default is 5

# Loading inputs & outputs

## Libraries

```python
import sklearn.neighbors as nei
import pandas as pd
import sklearn.model_selection as mod
```

## Data

```python
df = pd.read_csv("https://raw.githubusercontent.com/illinois-cse/data-fa14/gh-pages/data/iris.csv")
```

```python
df
```

## Visualise

```python
import seaborn as sns
sns.pairplot(df, hue="species")
```

## Inputs and Outputs

```python
inputs = df[['sepal_length','sepal_width','petal_length','petal_width']]
outputs = df['species']
# outputs
```

# Using K Nearest Neighbours and making predictions in SKLearn video

## Classifier

```python
knn = nei.KNeighborsClassifier(n_neighbors=5) ## this has instanciated knn model, not trained at this point in time
```

## Fit > training classifier

```python
knn.fit(inputs, outputs) # remembering inputs and outputs
```

## Predict

```python
df.loc[0]
```

```python
knn.predict([[5.1, 3.5, 1.4, 0.2]])
```

```python
knn.predict([[5.1, 3.5, 1.4, 0.2],[5.1, 3.5, 1.4, 0.2],[5.1, 3.5, 1.4, 0.2]])
```

# Testing the Classifier

# Evaluate

```python
knn.predict(inputs) ## getting some wrong, so we need to evaluate its performance
```

```python
knn.predict(inputs)==outputs
```

```python
(knn.predict(inputs)==outputs).sum() ## gives output of 145 True values, so 5 mistakes out of 150
```

```python
# SO we need to break up into trinaing set and test set (often train on more than test on)
# If you spit trainiing into first 2/3 and test on last 1/3 for setiso would be completely wrong
# Trained on 2 flowers and tested on 1
```

```python
mod.train_test_split(inputs, outputs, test_size=0.33)
```
