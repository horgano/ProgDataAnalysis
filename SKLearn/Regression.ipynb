## Straight Line data (noisy)

Lets look at our straight line data again

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
rng =  np.random.default_rng()
x = rng.random(10) * 10
y = 3*x + 4 + ((rng.random(10)*20) -10)    #Equation of a line pluus noise
plt.scatter(x,y)
plt.show
```

Reshape the data for passing it to the LinearRegression object

```python
x
```

```python
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
```

```python
x
```

Create an empty LinearRegression object and fit the model

```python
lr = LinearRegression()
lr.fit(x, y)
```

Make a prediction

```python
pred = lr.predict([[5]])
pred
```

Draw a line to represent the model

```python
x_test = np.linspace(0,20)
x_testShaped = x_test.reshape(-1,1)
```

```python
y_testPred = lr.predict(x_testShaped)
```

```python
plt.scatter(x,y)
plt.plot(x_test, y_testPred, 'r')
plt.legend(["Observed data", "Predicted line"])
plt.show()
```

# Iris data regression

```python
df = pd.read_csv("https://raw.githubusercontent.com/illinois-cse/data-fa14/gh-pages/data/iris.csv")
import seaborn as sns
sns.pairplot(df, hue="species")
```

```python
sW = df['sepal_width']. to_numpy()
sL = df['sepal_length']. to_numpy()
sW = sW.reshape(-1,1)
SL = sL.reshape(-1,1)
```

```python
lr = LinearRegression()
lr.fit(sW, sL)
```

Daw a line to represent the model

```python
sW_test = np.linspace(0,6)
sW_testShaped = sW_test.reshape(-1,1)
sL_testPred = lr.predict(sW_testShaped)
plt.scatter(sW, sL)
plt.plot(sW_test, sL_testPred, 'r')
plt.legend(["Predicted line", "Observed data"])
plt. show()
```
